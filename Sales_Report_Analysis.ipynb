```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('/content/sales_data.scv.csv',encoding='ISO_8859_1')
df
```

```python
df.info('unnamed1')
```

```python
df.drop(['unnamed1','Status'],axis=1,inplace=True)
df
```

```python
df['Cust_name'].unique()
```

```python
plt.figure(figsize=(10,5),facecolor='khaki')
dff=df.groupby(['Gender'])['Amount'].sum().reset_index()
dff
a=sns.barplot(x='Gender',y='Amount',data=dff,hue='Gender')
for i in a.containers:
  a.bar_label(i)
plt.title('SHOPPING-AMOUNT',color='black')
```

We can analyse  from sales data,Female's shopping amount is more than male.

```python
plt.figure(figsize=(20,7),facecolor='orange')
dff=df.groupby(['Occupation'])['Amount'].sum().reset_index()
dff
s=sns.barplot(x='Occupation',y='Amount',data=dff,hue='Occupation')
for i in s.containers:
  s.bar_label(i)
plt.title('SHOPPING-AMOUNT',color='black')
```

From the analysis, the IT sector records the highest shopping amount, approximately 1.47 × 10⁷. Other high-spending occupations include Healthcare (1.30 × 10⁷), Aviation (1.26 × 10⁷), and Banking (1.07 × 10⁷). On the lower end, Agriculture and Textile have the least shopping amounts, around 2.59 × 10⁶ and 3.20 × 10⁶, respectively.

```python
plt.figure(figsize=(20,7),facecolor='lime')
dff=df.groupby(['State'])['Amount'].sum().reset_index()
dff
s=sns.barplot(x='State',y='Amount',data=dff,hue='State')
for i in s.containers:
  s.bar_label(i)
plt.title('SHOPPING-AMOUNT',color='black')
```

The highest shopping amount appears in Uttar Pradesh (approximately 1.94 × 10⁷), followed by Maharashtra (1.44 × 10⁷) and Karnataka (1.35 × 10⁷). Delhi also has a significant value (1.16 × 10⁷), while states like Punjab, Rajasthan, and Telangana have relatively lower amounts

```python
plt.figure(figsize=(13,8),facecolor='grey')
dff=df.groupby(['Zone'])['Amount'].sum().reset_index()
dff
plt.pie(dff['Amount'],labels =dff['Zone'],explode=[0.2,0,0,0,0],shadow=True,startangle=90,autopct="%0.2f%%")
plt.show()
```

The Central region holds the largest shopping amount at 39.15%, followed by the Southern region with 25.04%. The Western region accounts for 17.29%, while the Eastern and another segment hold 6.63% and 11.88%, respectively

```python
plt.figure(figsize=(30,10),facecolor='violet')
dff=df.groupby(['Product_Category'])['Amount'].sum().reset_index()
dff
s=sns.barplot(x='Product_Category',y='Amount',data=dff,hue='Product_Category')
for i in s.containers:
  s.bar_label(i)
plt.title('SHOPPING-AMOUNT',color='black')
```

From the analysis,Highest amount spent in food followed by footwear and shoes or clothing and apparel, while product category tike office and veterinary and pet care have relatively spent lower amounts.



```python
plt.figure(figsize=(10,5),facecolor='gold')
dff=df.groupby(['Age Group'])['Orders'].sum().reset_index()
dff
s=sns.barplot(x='Age Group',y='Orders',data=dff,hue='Age Group')
for i in s.containers:
  s.bar_label(i)
plt.title('SHOPPING-AMOUNT',color='black')
```

From the analysis,Highest orders in the age group(26-35) followed by the age group (36-45) and (18-25), while age group tike(0-17)and(55+) have relatively  lower orders.

```python
plt.figure(figsize=(10,5),facecolor='brown')
dff=df.groupby(['Marital_Status'])['Orders'].sum().reset_index()
dff
s=sns.barplot(x='Marital_Status',y='Orders',data=dff,hue='Marital_Status')
for i in s.containers:
  s.bar_label(i)
plt.title('ORDER-COUNT',color='black')
```

From the analysis,Unmarried person order highest product as compared to married person.
